```python
noshoes=[]
boots=[]
flipflops=[]
loafers=[]
sandals=[]
soccer_shoes=[]
Y=[]
X=[]
shape=(140,140)
root_dir1="C:/Users/Dell/Desktop/topcoder/"#change this to your directory containing the training and testing images
for x in os.listdir(root_dir1+"noshoes/"):
    X.append(cv.resize(cv.imread(root_dir1+"noshoes/"+x),shape)/255)
    Y.append([1,0,0,0,0,0,0])
    
for x in os.listdir(root_dir1+"boots/"):
    X.append(cv.resize(cv.imread(root_dir1+"boots/"+x),shape)/255)
    Y.append([0,1,0,0,0,0,0])
for x in os.listdir(root_dir1+"flipflops/"):
    X.append(cv.resize(cv.imread(root_dir1+"flipflops/"+x),shape)/255)
    Y.append([0,0,1,0,0,0,0])
for x in os.listdir(root_dir1+"loafers/"):
    X.append(cv.resize(cv.imread(root_dir1+"loafers/"+x),shape)/255)
    Y.append([0,0,0,1,0,0,0])
for x in os.listdir(root_dir1+"sandals/"):
    X.append(cv.resize(cv.imread(root_dir1+"sandals/"+x),shape)/255)
    Y.append([0,0,0,0,1,0,0])
for x in os.listdir(root_dir1+"soccer_shoes/"):
    X.append(cv.resize(cv.imread(root_dir1+"soccer_shoes/"+x),shape)/255)
    Y.append([0,0,0,0,0,1,0])
for x in os.listdir(root_dir1+"Sneakers/"):
    X.append(cv.resize(cv.imread(root_dir1+"Sneakers/"+x),shape)/255)
    Y.append([0,0,0,0,0,0,1])    
```

```python
'''Perform train test split to use for valiadation'''
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
'''Using Pretrained model Densenet121'''
from tensorflow import keras
import tensorflow as tf
from keras.models import Sequential,Input,Model
from keras.layers import Conv2D, UpSampling2D, MaxPooling2D,Dense,Input,Dropout,Conv2DTranspose,concatenate
base_model =keras.applications.DenseNet121(
    include_top=False,
    weights="imagenet",
    input_shape=(140,140,3)
)
print(len(base_model.layers))
for x in base_model.layers[0:len(base_model.layers)]:
    x.trainable=False
    #base_model.layers[174].trainable=False
'''Unfreeze the Last Three Layers of the model'''    
base_model.layers[426].trainable=True
base_model.layers[425].trainable=True
base_model.layers[424].trainable=True
from keras.preprocessing.image import ImageDataGenerator
#Data Augmentation for generating more images
datagen = ImageDataGenerator(
    rotation_range = 30, zoom_range = 0.20, 
fill_mode = "nearest", shear_range = 0.20, horizontal_flip = True, 
width_shift_range = 0.1, height_shift_range = 0.2 )
datagen1 = ImageDataGenerator(
    rotation_range = 30, zoom_range = 0.20, 
fill_mode = "nearest", shear_range = 0.20, horizontal_flip = True, 
width_shift_range = 0.1, height_shift_range = 0.2)
inputs = keras.Input(shape=(140,140,3))
x = base_model(inputs,training=False)
x = keras.layers.GlobalAveragePooling2D()(x)
outputs = keras.layers.Dense(7,activation="softmax")(x)
model12 = keras.Model(inputs, outputs)
loss = tf.keras.losses.CategoricalCrossentropy() 
model12.compile(loss=loss,optimizer=keras.optimizers.SGD(0.5), metrics=['Recall','Precision'])
model12.summary()
```

```python
'''Using pretrained model DenseNet169'''
from tensorflow import keras
import tensorflow as tf
from keras.models import Sequential,Input,Model
from keras.layers import Conv2D, UpSampling2D, MaxPooling2D,Dense,Input,Dropout,Conv2DTranspose,concatenate
base_model =keras.applications.DenseNet169(
    include_top=False,
    weights="imagenet",
    input_shape=(140,140,3)
)
print(len(base_model.layers))
for x in base_model.layers[0:len(base_model.layers)]:
    x.trainable=False
    #base_model.layers[174].trainable=False
'''Unfreeze the Last Three Layers of the model'''    
base_model.layers[594].trainable=True
base_model.layers[593].trainable=True
base_model.layers[592].trainable=True

from keras.preprocessing.image import ImageDataGenerator
#Data Augmentation for generating more images
datagen = ImageDataGenerator(
    rotation_range = 30, zoom_range = 0.20, 
fill_mode = "nearest", shear_range = 0.20, horizontal_flip = True, 
width_shift_range = 0.1, height_shift_range = 0.2 )
datagen1 = ImageDataGenerator(
    rotation_range = 30, zoom_range = 0.20, 
fill_mode = "nearest", shear_range = 0.20, horizontal_flip = True, 
width_shift_range = 0.1, height_shift_range = 0.2)
inputs = keras.Input(shape=(140,140,3))
x = base_model(inputs,training=False)
x = keras.layers.GlobalAveragePooling2D()(x)
outputs = keras.layers.Dense(7,activation="softmax")(x)
model = keras.Model(inputs, outputs)
loss = tf.keras.losses.CategoricalCrossentropy() 
model.compile(loss=loss,optimizer=keras.optimizers.SGD(0.5), metrics=['Recall','Precision'])
model.summary()
```

```python
#Perform training on different deep learning models
datagen.fit(np.array(X_train))
datagen1.fit(np.array(X_test))
callback = keras.callbacks.EarlyStopping(monitor='loss', patience=10)
#model.fit(np.array(X_train),np.array(y_train),batch_size=32,epochs=5,validation_split=0.33)
model.fit(datagen.flow(np.array(X_train), np.array(y_train), batch_size=32),epochs=100,validation_data=datagen1.flow(np.array(X_test),np.array(y_test)),callbacks=[callback])
```
